```python
import pandas as pd;
data = pd.read_csv('c:/Bani_python/Bank Customer Churn Prediction.csv')
```

```python
data
```

```python
####Important columns in this data is below
# age,balance,active_member,gender,product_number,credit_score,tenure

#age → Different age groups show different loyalty and switching behavior.

#balance→ Customers with higher balances are usually more valuable and behave differently when deciding to leave.

#active_member→ Strong indicator of engagement; inactive customers are much more likely to churn.

#gender→ Can capture behavioral differences in churn patterns across groups.

#products_number→ More products = stronger relationship with the company → lower churn.

#credit_score→ Reflects financial stability; lower scores may correlate with higher churn risk.

#tenure→ Shows how long the customer has stayed; newer customers tend to churn more than long-term one


####Less important columns
# estimated_salary,country,credit_card,customer_id
#estimated_salary→ Usually doesn’t strongly influence whether a customer leaves. High or low salary customers churn at similar rates in many datasets.

#country→ Can matter slightly, but often not a strong direct driver of churn unless there are big regional differences. Its impact is usually weaker than behavioral features.

#credit_card→ Whether a customer has a credit card or not often shows little variation in churn behavior, so it adds limited predictive value.

#customer_id→ Purely an identifier (unique for each customer). It contains no meaningful pattern or relationship with churn, so it’s not useful for prediction at all.
```

```python
#As we compare the graph age is the one of the factor to effect the data 
#churn curved peaks around 45-55 age where stay curved peaks around 30-40 
#Customers aged 45–55 are more likely to churn compared to younger custom


import matplotlib.pyplot as plt
import seaborn as sns

churn=data[data['churn']==1]
stay=data[data['churn']==0]

sns.kdeplot(churn['age'])
sns.kdeplot(stay['age'])

plt.legend(['churn','stay'])
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Age Distribution: Churn vs Stay')
```

```python
# As we can see in this Estimated_salary is 


sns.kdeplot(x='estimated_salary',data=stay,color='red')
sns.kdeplot(x='estimated_salary',data=churn,color='green')
plt.legend(['stay','churn'])
plt.xlabel('Estimated_salary')
plt.ylabel('Density')
plt.title('Estimated salary: churn vs stay')
```

```python
ax=sns.countplot(data=data,x='country',hue='churn',palette=({0:'green',1:"red"}))

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(x='credit_card',data=data,hue='churn',palette=({0:'blue',1:"red"}))
for bars in ax.containers:
    ax.bar_label(bars)
```
